```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("labeled_data.csv")
print(data.shape)
```

```python
#fix_labels
data.loc[data["sentiment"] == 2, "sentiment"] = 4
data.loc[data["sentiment"] == 1, "sentiment"] = 5
#
#
#
print(data["sentiment"].unique())
```

```python
data.loc[data["sentiment"] == 5, "sentiment"] = 2
data.loc[data["sentiment"] == 3, "sentiment"] = 1
data.loc[data["sentiment"] == 4, "sentiment"] = 3
print(data["sentiment"].unique())
```

```python
data.to_csv("tweets_labeled.csv")
```

```python
negc = 439
neuc = 445
posc = 329
irrc = 1018
print(negc + neuc + posc + irrc)
```

```python
labels = ["Negative", "Neutral", "Positive", "Irrelevant"]
label_num = ["Negative\n(439)", "Neutral\n(445)", "Positive\n(329)", "Irrelevant\n(1018)"]
values = [negc, neuc, posc, irrc]

plt.pie(values, labels = label_num)
plt.title("Distrubution of Rows based on Classes")
plt.show()
```

```python
fig = plt.figure(figsize = (5, 5))
plt.bar(labels, values, width = 0.3) 
plt.xlabel("Classes")
plt.ylabel("Row Count")
plt.title("Number of Rows based on Classes")
plt.show()
```
